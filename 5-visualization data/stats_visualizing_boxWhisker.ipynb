# 박스 플롯 (Box and Whisker Plots / Box Plot)

박스 플롯(상자 수염 그림)은 데이터의 분포를 시각적으로 요약하여 보여주는 그래프이다. 데이터의 중앙값, 사분위수 범위, 그리고 이상치 등을 한눈에 파악하는 데 유용하다.

## 박스 플롯의 주요 구성 요소

* **중앙값 (Median)**:
    * 상자(box) 내부의 선으로 표시된다.
    * 전체 데이터를 크기순으로 정렬했을 때 정확히 가운데에 위치하는 값으로, 데이터의 50% 지점을 나타낸다.

박스 플롯을 통해 데이터의 대칭성, 퍼진 정도, 이상치의 유무 등을 빠르게 파악할 수 있으며, 여러 그룹 간의 데이터 분포를 비교하는 데도 효과적으로 사용된다.

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## create data for the bar plot
# 이전 bar plot에서 나온 난수 데이터 사용

# data sizes
m = 30 # rows
n =  6 # columns

# generate data
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
# now for the boxplot

plt.boxplot(data)
plt.show()
```

주황선은 중위값(median)을 나타내는 선이며, 전체 데이터를 크기순으로 정렬했을때, 정확이 가운데에 위치해 있는 값이다.

```python
# now with seaborn
sns.boxplot(data=data,orient='v')
plt.show()
```

orient 파라미터는 boxplot의 방향을 결정합니다. 즉, 그래프를 세로로 그릴지 가로로 그릴지를 정하는 역할을 합니다.

orient='v' (기본값): 수직 방향으로 boxplot을 그립니다. 이 경우, x축에는 범주형 변수가, y축에는 수치형 변수가 위치하게 됩니다.
orient='h': 수평 방향으로 boxplot을 그립니다. 이 경우, y축에는 범주형 변수가, x축에는 수치형 변수가 위치하게 됩니다.
따라서, orient는 **"boxplot의 방향 (수직 또는 수평)을 지정하는 파라미터"**

```python
# or as a pandas data frame
df = pd.DataFrame(data,columns=['zero','one','two','three','four','five'])
sns.boxplot(data=df,orient='h')
plt.show()
```

### 에러바
값이 얼마나 정확한지, 변동성이 얼마나 큰지 나타내고 싶을때 사용

- 에러 바는 막대 위나 아래에 짧은 선으로 표시되어서, 데이터의 표준편차(standard deviation), 표준오차(standard error), 혹은 신뢰구간(confidence interval) 등을 나타낼 수 있어요.

    - 표준편차는 데이터가 평균으로부터 얼마나 흩어져 있는지를 보여주는 값이에요. 키로 비유하자면, 우리 반 친구들의 키가 다 비슷비슷한지, 아니면 키 큰 친구와 키 작은 친구가 골고루 섞여 있는지를 알려주는 거죠. 표준편차가 크면 에러 바의 길이가 길어지고, 작으면 짧아져요.
    - 표준오차는 샘플(표본)에서 계산한 평균값이 실제 모집단의 평균값과 얼마나 차이가 날지를 추정한 값이에요. 우리가 전체 학생의 키를 다 잴 수 없어서, 한 반 학생들의 키만 재서 평균을 냈을 때, 이 평균이 전체 학생 평균과 얼마나 비슷할지를 나타내는 거죠.
    - 신뢰구간은 실제 모집단의 평균값이 존재할 것이라고 예상되는 범위를 나타내요. "95% 신뢰구간"이라고 하면, 우리가 샘플 조사를 100번 반복했을 때 95번은 실제 평균값이 이 구간 안에 들어올 것이라고 기대하는 거예요.

## 다양한 형태로 막대 그래프 활용하기

1. pandas DataFrame으로 그룹형 막대 그래프 그리기:

pandas는 파이썬에서 데이터 분석을 할 때 아주 강력한 라이브러리예요. 특히 표(table) 형태의 데이터를 다룰 때 유용한 DataFrame이라는 자료구조를 제공하죠.
DataFrame을 사용하면 여러 그룹의 데이터를 한 번에 막대 그래프로 표현하기가 아주 쉬워요. 예를 들어, 여러 반 학생들의 국어, 영어, 수학 점수를 비교하는 그룹형 막대 그래프를 손쉽게 그릴 수 있답니다. DataFrame의 plot(kind='bar')

2. 데이터 관점 바꾸기: 전치(Transpose)

때로는 데이터를 다른 관점에서 보고 싶을 때가 있어요. 예를 들어, 처음에는 '학생'별로 '과목' 점수를 봤다면, 이번에는 '과목'별로 '학생' 점수를 보고 싶을 수 있죠.
이럴 때 사용하는 것이 바로 **전치(transpose)**예요. 행과 열을 서로 맞바꾸는 거죠. pandas DataFrame에서는 .T를 붙여서 간단하게 데이터를 전치할 수 있어요. 전치된 DataFrame으로 막대 그래프를 그리면 또 다른 인사이트를 얻을 수 있답니다.

3. 여러 그래프 한 번에 보기: plt.subplots()

여러 개의 연관된 그래프를 한 화면에 나란히 놓고 비교하고 싶을 때가 있죠? 이럴 때 plt.subplots() 함수가 아주 유용해요.
fig, ax = plt.subplots(nrows=2, ncols=2)처럼 사용하면, 2x2 형태로 총 4개의 하위 그래프(subplot)를 그릴 수 있는 공간을 만들 수 있어요. 그리고 ax[0, 0], ax[0, 1]처럼 각 위치를 지정해서 그래프를 그릴 수 있답니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

index_names = ['alpha', 'beta']
column_names = ['1Q', '2Q', '3Q', '4Q']
data = [[80, 85, 90, 75],
        [70, 95, 80, 90]] 
df = pd.DataFrame(data, index=index_names, columns=column_names)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(12,5))

df.plot(kind='bar', ax=ax[0], title='score per team', xlabel='team', ylabel='score')
df.T.plot(kind='bar', ax=ax[1], title='Score per Quarter', xlabel='Quarter', ylabel='Score')

```
